```python
from sklearn.cluster import KMeans
```

```python
# k= 2 clusters
kmeans=KMeans(n_clusters=2)
```

```python
# fit the cluster
kmeans.fit(df.drop('Private',axis=1))
```

```python
# center the cluster
kmeans.cluster_centers_
```

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```

```python

inertia = []
K_range = range(1, 11)

for k in K_range:
    km = KMeans(n_clusters=k)
    km.fit(df.drop('Private', axis=1))
    inertia.append(km.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

This graph utilizes the elbow plot method to help you measure how tightly packed each cluster is.

From observing the graph, we can see that K1 and K2 have a sharp drop, but as it continues adding clusters, the the drops get lower and less sharp.


```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(df.drop('Private', axis=1))

# Check cluster sizes
pd.Series(kmeans.labels_).value_counts()
```

```python
from sklearn.metrics import silhouette_score
scores = []
K_range = range(2, 11)  # silhouette needs at least 2 clusters

for k in K_range:
    km = KMeans(n_clusters=k)
    labels = km.fit_predict(df.drop('Private', axis=1))
    scores.append(silhouette_score(df.drop('Private', axis=1), labels))

plt.figure(figsize=(10, 6))
plt.plot(K_range, scores, 'ro-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis')
plt.show()
```

This graph gives each point a score from -1 to 1 to determine whether its closer to its own cluster center or the neighboring cluster's center. 

From looking at the data, 2 clusters is farthest from zero, but as you add more, its gets closer to zero meaning the clusters are overlapping.

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(df.drop('Private', axis=1))

# Check cluster sizes
pd.Series(kmeans.labels_).value_counts()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, k in enumerate([2, 3, 4]):
    km = KMeans(n_clusters=k)
    km.fit(df.drop('Private', axis=1))
    
    axes[i].scatter(df['Outstate'], df['F.Undergrad'], 
                    c=km.labels_, cmap='coolwarm', alpha=0.6, s=30)
    axes[i].set_title(f'K = {k}')
    axes[i].set_xlabel('Outstate')
    axes[i].set_ylabel('F.Undergrad')

plt.tight_layout()
plt.show()
```

This graph takes two features and plots every school as different colored dots that show which cluster K-Means its assigned to.

By looking at the graph, we can see that K=2 has more clean splits while K3 and K4 have more mixed colors meaning that the features don't define the cluster boundary well.

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df.drop('Private', axis=1))

# Check cluster sizes
pd.Series(kmeans.labels_).value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(df.drop('Private', axis=1))

# Compare unscaled vs scaled
km_raw = KMeans(n_clusters=2).fit(df.drop('Private', axis=1))
km_scaled = KMeans(n_clusters=2).fit(scaled)

print("Unscaled accuracy:")
print(confusion_matrix(df['Cluster'], km_raw.labels_))

print("\nScaled accuracy:")
print(confusion_matrix(df['Cluster'], km_scaled.labels_))
```

This isn't a graph but a matrix to help compare labels against the cluster assignments while counting the matches and mismatches.

Looking at the diagnoal numebrs, we can see the correct assignments while the off-diagnals are mistakes.


```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(df.drop('Private', axis=1))

# Check cluster sizes
pd.Series(kmeans.labels_).value_counts()
```
